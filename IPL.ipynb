```python
#Your task is to design a dashboard containing: 
#1. Top 10 Batsman: 
  #a.based on Runs scored
  #b.based on Number of 6s
  #c.based on Number of 4s
  #d.based to Number of 100s


```

```python
  #a.based on Runs scored
    
scores = pd.merge(ipl_bbb,ipl_pl, left_on="Striker_Id",right_on = "Player_Id")
```

```python
scores["Batsman_Scored"].value_counts()
```

```python
def clean_scores(score):
    try:
        return int(score)
    except:
        return float('nan')
```

```python
scores["Batsman_Scored"] = scores["Batsman_Scored"].apply(clean_scores)
```

```python
scores["Batsman_Scored"].value_counts()
```

```python
scores["Batsman_Scored"].groupby(scores.Player_Name).sum().nlargest(10)
```

```python
  #b.based on Number of 6s

scores[scores.Batsman_Scored==6][["Batsman_Scored","Player_Name"]].groupby(scores.Player_Name).count().nlargest(10,"Batsman_Scored")
```

```python
#c.based on Number of 4s

scores[scores.Batsman_Scored==4][["Batsman_Scored","Player_Name"]].groupby(scores.Player_Name).count().nlargest(10,"Batsman_Scored")
```

```python
#d.based to Number of 100s

individual_match_scores=scores[["Batsman_Scored","Player_Name"]].groupby([scores.Match_Id,scores.Player_Name]).sum()
individual_match_scores.head(20)
```

```python
individual_match_scores[individual_match_scores.Batsman_Scored>100].groupby("Player_Name").count().nlargest(10,"Batsman_Scored")
```
